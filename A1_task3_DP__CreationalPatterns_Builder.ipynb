## Task 3. Implementing **Builder** Pattern in Python - 10 points

1. Imagine that you need to develop a Customer Component to get customers details (say to store in a DB).

2. Customers are shown a form that accepts several fields: mandatory and optional



3. By using the Builder design pattern create an object of this Component step by step without passing all the values to the class constructor.

To do that first you will need to refine the UML class diagram below to develop 2 Customer Builders (WebAppBuilder and MobileAppBuilder) and a Director to manage the process of assembling the object accordingly the the following order: firstName, middleName, lastName, primaryEmail, secondaryEmail, primaryMobileNumber, secondaryMobileNumber
* WebAppBuilder has the following fields into the component object: firstName, middleName, lastName, primaryEmail, secondaryEmail, primaryMobileNumber, secondaryMobileNumber.
* MobileAppBuilder doesn't need middleName, secondaryEmail and secondaryMobileNumber
